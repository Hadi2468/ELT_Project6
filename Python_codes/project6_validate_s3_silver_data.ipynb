# Project 6: Validation of S3 Silver Layer

```python
import pandas as pd
import glob
```

## fact_marketing

```python
# List all Parquet files in folder
files = glob.glob("../data/Silver/fact_marketing/*/*.parquet")
# display(files)

# Read and concatenate them
df_fact_marketing = pd.concat([pd.read_parquet(f, engine="fastparquet") for f in files], ignore_index=True)
display(df_fact_marketing.head())
df_fact_marketing.info()
```

## fact_events

```python
# List all Parquet files in folder
files = glob.glob("../data/Silver/fact_events/*/*.parquet")
# display(files)

# Read and concatenate them
df_fact_events = pd.concat([pd.read_parquet(f, engine="fastparquet") for f in files], ignore_index=True)
display(df_fact_events.tail())
df_fact_events.info()
```
